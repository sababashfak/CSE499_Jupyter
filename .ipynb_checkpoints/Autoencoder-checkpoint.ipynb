```python
import os
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split

# Folder paths
input_dir = "F:/Education/NSU/CSE/CSE499/Implementation/Image Data/plots"
output_dir = "F:/Education/NSU/CSE/CSE499/Implementation/Image Data/matrix"

input_size = (400, 1000)  # (width, height)
output_size = (50, 50)

X_data = []
Y_data = []

# Loop over possible X values from 1 to 1500
for i in range(1, 1501):
    input_path = os.path.join(input_dir, f"graph_{i}.png")
    output_path = os.path.join(output_dir, f"{i}.png")

    # Check if both input and output files exist
    if os.path.exists(input_path) and os.path.exists(output_path):
        # Load and preprocess input image
        in_img = Image.open(input_path).convert("L").resize(input_size)
        in_arr = np.array(in_img) / 255.0
        X_data.append(in_arr)

        # Load and preprocess output image
        out_img = Image.open(output_path).convert("L").resize(output_size)
        out_arr = np.array(out_img) / 255.0
        Y_data.append(out_arr)
    else:
        # Skipping this pair because one of them is missing
        continue

# Convert to NumPy arrays and reshape for CNN
X_data = np.array(X_data).reshape(-1, 1000, 400, 1)
Y_data = np.array(Y_data).reshape(-1, 50, 50, 1)

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42)

print(f"✅ Total usable pairs found: {len(X_data)}")
```

```python
import matplotlib.pyplot as plt

# Number of samples to visualize
num_samples = 10

plt.figure(figsize=(10, 4))

for i in range(num_samples):
    # Input image: 1000x400
    plt.subplot(2, num_samples, i + 1)
    plt.imshow(X_data[i].reshape(1000, 400), cmap='gray')
    plt.title(f"Input {i}")
    plt.axis('off')

    # Output image: 50x50
    plt.subplot(2, num_samples, i + 1 + num_samples)
    plt.imshow(Y_data[i].reshape(50, 50), cmap='gray')
    plt.title(f"Output {i}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Resizing
from tensorflow.keras.optimizers import Adam
```

```python
input_img = Input(shape=(1000, 400, 1))

# ----- Encoder -----
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)  # 500×200×32

x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((5, 4), padding='same')(x)  # 100×50×64

x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)  # 50×25×128

# ----- Decoder -----
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)  # 100×50×128

x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((5, 4))(x)  # 500×200×64

x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)  # 1000×400×32

# Final 1-channel image
x = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)  # still 1000×400×1

# Resize to match label size
output_img = Resizing(50, 50)(x)  # Final: 50×50×1

```

```python
autoencoder = Model(input_img, output_img)
autoencoder.compile(optimizer=Adam(1e-4), loss='mse')
autoencoder.summary()

```

```python
autoencoder.fit(
    X_train, Y_train,
    epochs=50,
    batch_size=16,
    shuffle=True,
    validation_data=(X_test, Y_test)
)
```

```python
predicted = autoencoder.predict(X_test)
predicted_images = predicted.reshape((-1, 50, 50))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(Y_test[i].reshape(50, 50), cmap='gray')
    plt.title("Original")
    
    plt.subplot(2, 5, i + 6)
    plt.imshow(predicted_images[i], cmap='gray')
    plt.title("Generated")

plt.tight_layout()
plt.show()
```
