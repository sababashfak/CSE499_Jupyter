```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
```

```python
input_dir = 'F:/Education/NSU/CSE/CSE499/Implementation/Dataset/combined_output.csv'
x_axis_dir = 'F:/Education/NSU/CSE/CSE499/Implementation/column1.csv'
df1 = pd.read_csv(input_dir, header=None, low_memory=False)
df2 = pd.read_csv(x_axis_dir)
```

```python
x_axis = df2.values.flatten()
x_axis
```

```python
output_dir = "F:/Education/NSU/CSE/CSE499/Implementation/Dataset/plots"
os.makedirs(output_dir, exist_ok=True)
```

```python
# Indices for line1 and line2
line1_cols = list(range(26, 2028, 2))  # even columns
line2_cols = list(range(27, 2028, 2))  # odd columns
#line1_cols
line2_cols
```

```python
# Sanity check (should be 1001 each)
assert len(line1_cols) == 1001 and len(line2_cols) == 1001
```

```python
#lOOP

for i in range(len(df1)-1):
    line1 = df1.iloc[i+1, line1_cols].values.astype(float)
    line2 = df1.iloc[i+1, line2_cols].values.astype(float)
    print(line2)

    plt.figure(figsize=(10, 4))
    plt.plot(x_axis, line1, label="Real Line", color='blue')
    plt.plot(x_axis, line2, label="Imaginary Line", color='red', linestyle='--')
    plt.title(f"Graph {i+1}")
    plt.xlabel("Frequency (GHz)")
    plt.ylabel("Reflection Coefficient")
    plt.ylim(-1, 1)  # <- Set Y-axis range
    plt.legend()
    plt.grid(True)

    # Save the figure
    plt.savefig(f"{output_dir}/graph_{i+1}.png")
    plt.close()

print(" All graphs with interleaved lines have been saved.")
```
